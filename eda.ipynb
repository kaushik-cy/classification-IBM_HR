```python
import pandas as pd
df = pd.read_csv(r'data/dataset_e1.csv')
df.head()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

for col in df.columns:
    plt.figure(figsize = (16,4))
    sns.histplot(df[col])
    plt.show()
```

```python
plt.subplots(figsize = (16,5))
feature = 'Age'
df[df['Attrition'] == 0][feature].plot.kde(label = feature+' Attrition = 0')
df[df['Attrition'] == 1][feature].plot.kde(label = feature+' Attrition = 1')
plt.legend()
```
